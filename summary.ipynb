# Summary

Comparing the models it is clear that: **EfficientDet_D1 performs best**:
* **Total Loss of 0.385** (much better than "ssd_mobilenet_v2_fpnlite" and 3-times as good as "ssd_resnet50_v1_fpn")
* **Classification Loss of 0.33** (better than "ssd_mobilenet_v2_fpnlite" at 0.4 and much better than "ssd_resnet50_v1_fpn" at 0.53)
* **mAP (mean Average Performance) of 0.07708** (better than "ssd_mobilenet_v2_fpnlite" and twice as good as "ssd_resnet50_v1_fpn")
